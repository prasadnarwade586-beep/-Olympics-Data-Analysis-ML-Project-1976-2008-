```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/olympics_cleaned.csv")
```

1.Weighted medal trend

```python
df.groupby("Year")["Medal_Points"].sum().plot(
    figsize=(10,5), title="Weighted Olympic Medal Trend"
)
plt.show()
```

2.Country dominance

```python
country_share = (
    df.groupby("Country")["Medal_Points"].sum()
    / df["Medal_Points"].sum()
) * 100

country_share.sort_values(ascending=False).head(10)
```

3.Sport dominance

```python
df.groupby(["Sport", "Country"])["Medal_Points"].sum() \
  .sort_values(ascending=False) \
  .groupby(level=0).head(1)
```
