```python
import pandas as pd

df = pd.read_csv("../data/raw/Summer-Olympic-medals-1976-to-2008.csv", encoding="latin1")
```

```python
df = df.drop(columns=["Event_gender", "Country_Code"])
```

```python
df = df.dropna(subset=["Medal", "Athlete", "Sport"])
```

```python
df["Year"] = df["Year"].astype(int)
```

#Medal weight

```python
df["Medal_Points"] = df["Medal"].map({"Gold":3, "Silver":2, "Bronze":1})
```

#Athlete experience

```python
df["Athlete_Experience"] = df.groupby("Athlete")["Year"].transform("nunique")
```

#Country strength

```python
df["Country_Strength"] = df.groupby("Country")["Medal"].transform("count")
```

```python
df.to_csv("../data/processed/olympics_cleaned.csv", index=False)
```
